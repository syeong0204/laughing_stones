```python
import pandas as pd
import numpy as np
```

```python
USA_Music_csv = "csv/US.csv"
albumList_csv = "csv/albumlist.csv"
```

```python
usa_df = pd.read_csv(USA_Music_csv)
albumList_df = pd.read_csv(albumList_csv, encoding='UTF-8')
```

```python
#albumList_df
usa_df.rename(columns = {'artists':'Artist'}, inplace = True)
usa_df.rename(columns = {'release_date':'Year'}, inplace = True)
usa_df.rename(columns = {'popularity':'Popularity'}, inplace = True)
usa_df.rename(columns = {'name':'SongTitle'}, inplace = True)
```

```python
usa_df
#albumList_df
```

```python
smaller_usa_df = usa_df.drop(columns=['id', 'explicit', 'id_artists', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature'])
smaller_usa_df
```

```python
smaller_usa_df['Artist'] = smaller_usa_df['Artist'].map(lambda x: x.lstrip('[\'').rstrip('aAbBcC'))
smaller_usa_df['Artist'] = smaller_usa_df['Artist'].map(lambda y: str(y)[:-2])

smaller_usa_df['Year'] = smaller_usa_df['Year'].map(lambda z: str(z)[:-6])

smaller_usa_df
```

```python
#albumList_df
modified_albumList_df = albumList_df.drop(columns=['Album', 'Subgenre'])
modified_albumList_df
```

```python
#smaller_usa_df & modified_albumList_df
#pd.to_numeric(smaller_usa_df['Year']).astype(int)

smaller_usa_df['Year'] = smaller_usa_df['Year'].fillna(0)
#smaller_usa_df.value_counts()
modified_albumList_df['Year'] = modified_albumList_df['Year'].fillna(0)

smaller_usa_df['Year'] = pd.to_numeric(smaller_usa_df['Year'], errors="coerce")
#smaller_usa_df
modified_albumList_df['Year'] = pd.to_numeric(modified_albumList_df['Year'], errors="coerce")
modified_albumList_df
```

```python
merged_music_df = pd.merge(smaller_usa_df, modified_albumList_df)
merged_music_df['Year'] = merged_music_df['Year'].astype(int)

merged_music_df
```

```python
#data['amount']=data['amount'].replace(0, np.nan)
#merged_music_df = merged_music_df.replace(0, pd.np.nan).dropna(axis=0, how='any', subset=columns).fillna(0)
merged_music_df['Year'] = merged_music_df['Year'].replace(0, np.nan)
merged_music_df.info()
```

```python
merged_music_df.to_csv('csv/merged_music_data.csv')
```
